```python
def parse_party(df):
    equals = []
    for c in df:
        equals.append(df[c].str.contains('=').dropna().astype(int).sum())
    equals_column = np.argmax(equals)    
    df = df.dropna(subset=[df.columns[equals_column]])
    abbrev_column = equals_column - 1
    name_column = equals_column + 1
    ret = df.iloc[:,[abbrev_column, name_column]]
    ret.columns = ['party', 'party_name']
    return ret.dropna().astype(str).applymap(str.strip)
```

```python
def parse_20XX_up_to_and_including_2012(xls_filename, year):
    cycle = str(year)
    
    party_labels_sheetname = f'{cycle} Party Labels'
    if cycle == '2004':
        results_sheetname = f'{cycle} US HOUSE & SENATE RESULTS'
    elif cycle in ['2006', '2010']:
        results_sheetname = f'{cycle} US House & Senate Results'
    elif cycle == '2008':
        results_sheetname = f'{cycle} House and Senate Results'
    elif cycle in ['2012']:
        results_sheetname = f'{cycle} US House & Senate Resuts'
    
    dfparty = pd.read_excel(xls_filename, sheet_name=party_labels_sheetname)
    dfresults = pd.read_excel(xls_filename, sheet_name=results_sheetname)
    
    # prepare dfparty
    ptydf = parse_party(dfparty)
    ptydf['cycle'] = cycle
    
    # prepare dfresults
    if cycle in ['2004', '2006']:
        FECID_colname = 'FEC ID'
    elif cycle in ['2008', '2010', '2012']:
        FECID_colname = 'FEC ID#'
        
    dfresults = dfresults.dropna(subset=[FECID_colname])
    dfresults[FECID_colname] = dfresults[FECID_colname].map(str).str.strip()
    dfresults = dfresults[~(dfresults[FECID_colname].map(len) == 0)]
        
    
    if cycle in ['2004', '2006']:
        dfresults = dfresults[[
            'STATE ABBREVIATION',
            'DISTRICT',
            'FEC ID',
            'INCUMBENT INDICATOR',
            'FIRST NAME',
            'LAST NAME',
            'LAST NAME, FIRST',
            'PARTY',
            'PRIMARY',
            'PRIMARY %',
            'RUNOFF',
            'RUNOFF %',
            'GENERAL',
            'GENERAL %',
            'GE RUNOFF',
            'GE RUNOFF %'
        ]]
        dfresults['GE WINNER IND'] = np.nan
        
    elif cycle in ['2008']:
        dfresults = dfresults[[
            'STATE ABBREVIATION',
            'DISTRICT',
            'FEC ID#',
            'INCUMBENT INDICATOR (I)',
            'CANDIDATE NAME (First)',
            'Candidate Name (Last)',
            'CANDIDATE NAME',
            'PARTY',
            'PRIMARY',
            'PRIMARY %',
            'RUNOFF',
            'RUNOFF %',
            'GENERAL ',  # yes. general space.
            'GENERAL %',
            'GE RUNOFF',
            'GE RUNOFF %'
        ]]
        dfresults['GE WINNER IND'] = np.nan
        
    elif cycle in ['2010']:
        dfresults = dfresults[[
            'STATE ABBREVIATION',
            'DISTRICT',
            'FEC ID#',
            'INCUMBENT INDICATOR (I)',
            'CANDIDATE NAME (First)',
            'CANDIDATE NAME (Last)',
            'CANDIDATE NAME (Last, First)',
            'PARTY',
            'PRIMARY',
            'PRIMARY %',
            'RUNOFF',
            'RUNOFF %',
            'GENERAL ',
            'GENERAL %',
        ]]
        dfresults['GE RUNOFF'] = np.nan
        dfresults['GE RUNOFF %'] = np.nan
        dfresults['GE WINNER IND'] = np.nan
        
    elif cycle in ['2012']:
        dfresults = dfresults[[
            'STATE ABBREVIATION',
            'D',
            'FEC ID#',
            '(I)',
            'CANDIDATE NAME (First)',
            'CANDIDATE NAME (Last)',
            'CANDIDATE NAME',
            'PARTY',
            'PRIMARY VOTES',
            'PRIMARY %',
            'RUNOFF VOTES',
            'RUNOFF %',
            'GENERAL VOTES ',
            'GENERAL %',
            'GE RUNOFF ELECTION VOTES (LA)',
            'GE RUNOFF ELECTION % (LA)',
            'GE WINNER INDICATOR',
        ]]
    
    dfresults.columns = [
        'state',
        'district',
        'fec_id',
        'incumbent',
        'firstname',
        'lastname',
        'lastfirst',
        'party',
        'primary_votes',
        'primary_pct',
        'runoff_votes',
        'runoff_pct',
        'general_votes',
        'general_pct',
        'ge_runoff_votes',
        'ge_runoff_pct',
        'ge_winner_ind'
    ]
    
    dfresults['chamber'] = np.nan  # forward-compatibility
    dfresults['cycle'] = cycle
    dfresults['party'] = dfresults['party'].astype(str).str.strip()
    
    if cycle == '2006':
        dfresults['party'] = dfresults['party'].map(lambda x: 'R' if x == 'REP' else 'D' if x == 'DEM' else x)
    
    return ptydf, dfresults
```

```python
def parse_20XX_2014_onwards(xls_filename, year):
    """
    In 2014 the certified election results workbook was split by FEC
    into two sheets: one for the House and one for the Senate.
    
    To manage this change and keep things relatively clean, this function
    handles parsing election results for 2014 and beyond.
    """
    
    cycle = str(year)
    if int(year) < 2014:
        raise Exception('you should use parse_20XX_up_to_and_including_2012() instead of this function')
    
    party_labels_sheetname = f'{cycle} Party Labels'
    if cycle in ['2014', '2016']:
        house_results_sheetname = f'{cycle} US House Results by State'
        senate_results_sheetname = f'{cycle} US Senate Results by State'
    
    dfparty = pd.read_excel(xls_filename, sheet_name=party_labels_sheetname)
    
    # prepare dfparty
    ptydf = parse_party(dfparty)
    ptydf['cycle'] = cycle
    
    # prepare dfresults
    for chambername, sheetname in zip(['house', 'senate'], [house_results_sheetname, senate_results_sheetname]):
        chamberresults = pd.read_excel(xls_filename, sheet_name=sheetname)
    
        if cycle in ['2014', '2016']:
            FECID_colname = 'FEC ID#'
        
        chamberresults = chamberresults.dropna(subset=[FECID_colname])
        chamberresults[FECID_colname] = chamberresults[FECID_colname].map(str).str.strip()
        chamberresults = chamberresults[~(chamberresults[FECID_colname].map(len) == 0)]

        if cycle in ['2014', '2016']:
            chamberresults = chamberresults[[
                'STATE ABBREVIATION',
                'D',
                'FEC ID#',
                '(I)',
                'CANDIDATE NAME (First)',
                'CANDIDATE NAME (Last)',
                'CANDIDATE NAME',
                'PARTY',
                'PRIMARY VOTES',
                'PRIMARY %',
                'RUNOFF VOTES',
                'RUNOFF %',
                'GENERAL VOTES ',
                'GENERAL %',
                'GE RUNOFF ELECTION VOTES (LA)',
                'GE RUNOFF ELECTION % (LA)',
                'GE WINNER INDICATOR'
            ]]
            chamberresults['chamber'] = chambername
    
        chamberresults.columns = [
            'state',
            'district',
            'fec_id',
            'incumbent',
            'firstname',
            'lastname',
            'lastfirst',
            'party',
            'primary_votes',
            'primary_pct',
            'runoff_votes',
            'runoff_pct',
            'general_votes',
            'general_pct',
            'ge_runoff_votes',
            'ge_runoff_pct',
            'ge_winner_ind',
            'chamber'
        ]
    
        chamberresults['cycle'] = cycle
        chamberresults['party'] = chamberresults['party'].astype(str).str.strip()
        
        if chambername == 'house':
            dfresults = chamberresults
        else:
            dfresults = pd.concat([dfresults, chamberresults], ignore_index=True)
    
    return ptydf, dfresults
```

```python
def clean_vote_col(col):
    col = col.fillna(-1)\
             .astype(str).str.replace(r'\#+', '-1', flags=re.IGNORECASE, regex=True)\
             .str.replace(r'unopposed', '9999999999', flags=re.IGNORECASE, regex=True)\
             .map(lambda x: int(float(x)))
    return col
```

```python
def guess_winner(df):
    cycle = df['cycle'].iloc[0]
    st = df['state'].iloc[0]
    dist = df['district'].iloc[0]
    
    if df['ge_runoff_votes'].notnull().sum() > 0:
        ge_runoff_votes = df['ge_runoff_votes'].copy()
        try:
            df['ge_runoff_votes'] = clean_vote_col(df['ge_runoff_votes'])
        except Exception as e:
            print(f'Error processing {cycle} RUNOFF election: {st}-{dist}', file=sys.stderr)
            raise e
        fecid_totals_runoff = df.groupby('fec_id')['ge_runoff_votes'].sum()
        max_ge_runoff_votes = fecid_totals_runoff.max()
        fecids_with_max_ge_runoff_votes = list(fecid_totals_runoff[fecid_totals_runoff == max_ge_runoff_votes].index)
        rows_with_max_ge_runoff_vote_fecids = df[df['fec_id'].isin(fecids_with_max_ge_runoff_votes)]
        if len(fecids_with_max_ge_runoff_votes) > 1 and max_ge_runoff_votes == -1:
            print(f'WARNING: {len(fecids_with_max_ge_runoff_votes)}-way tie with runoff votes count for {cycle} {st}-{dist}.')
            print(f'All vote tallies are -1, this is probably a special election primary. See results below:')
            display(HTML(rows_with_max_ge_runoff_vote_fecids.to_html()))
            display(HTML('<hr>'))
            df['ge_winner_ind_guess'] = 'SPECIAL_PRIMARY_NO_GENERAL'
        elif len(fecids_with_max_ge_runoff_votes) > 1:
            # Guess winner using runoff percentages
            # NOT IMPLEMENTED YET
            raise Exception(f'Unbreakable tie in runoff election: {cycle} {st}-{dist}')
        else:
            df['ge_winner_ind_guess'] = df['fec_id'].isin(fecids_with_max_ge_runoff_votes).map(lambda x: 'W' if x else np.nan)
        df['ge_runoff_votes'] = ge_runoff_votes
    
    else:
        # Guess winner using general votes count
        general_votes = df['general_votes'].copy()
        try:
            df['general_votes'] = clean_vote_col(df['general_votes'])
        except Exception as e:
            print(f'Error processing {cycle} GENERAL VOTING election: {st}-{dist}', file=sys.stderr)
            raise e
        fecid_totals = df.groupby('fec_id')['general_votes'].sum()
        max_general_votes = fecid_totals.max()
        fecids_with_max_votes = list(fecid_totals[fecid_totals == max_general_votes].index)
        rows_with_max_fecids = df[df['fec_id'].isin(fecids_with_max_votes)]

        if len(fecids_with_max_votes) > 1 and max_general_votes == -1:
            print(f'WARNING: {len(fecids_with_max_votes)}-way tie with general votes count for {cycle} {st}-{dist}:')
            print(f'All vote tallies are -1, this is probably a special election primary. See results below:')
            display(HTML(rows_with_max_fecids.to_html()))
            display(HTML('<hr>'))
            
            df['ge_winner_ind_guess'] = 'SPECIAL_PRIMARY_NO_GENERAL'
        elif len(fecids_with_max_votes) > 1:
            # Guess winner using general percentages
            # NOT IMPLEMENTED YET
            raise Exception(f'Unbreakable tie in general election: {cycle} {st}-{dist}')
        else:
            df['ge_winner_ind_guess'] = df['fec_id'].isin(fecids_with_max_votes).map(lambda x: 'W' if x else np.nan)
        df['general_votes'] = general_votes
    
    return df
    
    
    raise Exception('Unable to determine election winner.')
```

```python
r_fec_id = re.compile(r'/data/candidate/([A-Z0-9]{9})/')
def extract_fec_ids(electionjson):
    try:
        electionjson = eval(electionjson.replace('null', 'None').replace('false', 'False').replace('true', 'True'))
        elec = electionjson['current_election']
        ret = []
        for k, v in elec.items():
            fec_id = re.search(r_fec_id, v['fec_url']).group(1)
            name = v['candidate_name']
            ret.append((fec_id, name))
        return ret
    except:
        return []
```

```python
def lookup_cand_fecid(fecids, name, office, stateabbr):
    officepart = office[0]  # 'H' for 'House', 'S' for 'Senate'
    nameparts = name.lower().split(' ')
    possible_matches = []
    for pair in fecids:
        fecname = pair[1].lower()
        fecid = pair[0]
        if fecid[0] != officepart or fecid[2:4] != stateabbr:
            continue
        if all(map(lambda x: x in fecname, nameparts)):
            return fecid
        possible_matches.append(pair)
    if len(possible_matches) == 0:
        return None
    if len(possible_matches) == 1:
        return possible_matches[0][0]
    bestmatch = None
    bestmatch_score = 999
    for p in possible_matches:
        fecname = p[1].lower()
        fecnametokens = re.sub(r'[^A-Za-z\ ]', '', fecname).split(' ')
        nametokens = re.sub(r'[^A-Za-z\ ]', '', name.lower()).split(' ')
        score = 0
        if fecnametokens[0] in nameparts[:2]:
            score -= (5 + len(fecnametokens[0]))
        if nameparts[0] in fecnametokens[:2]:
            score -= (5 + len(nameparts[0]))
        if fecnametokens[-1] in nameparts[-2:]:
            score -= (20 + len(fecnametokens[-1]))
        if nameparts[-1] in fecnametokens[-2:]:
            score -= (20 + len(fecnametokens[-1]))
        for fn in fecnametokens[1:-1]:
            if fn in nameparts:
                score -= (10 + len(fn))
        score += len(set(name).symmetric_difference(set(fecname)))
        if score < bestmatch_score and score < 3:
            bestmatch = p[0]
            bestmatch_score = score
            
    return bestmatch
```
